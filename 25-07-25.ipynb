```python
#Grouping in pandas means organizing your data into groups based on columns .once grouped u can perform actions like dfinding the total ,average  and count 
# this acan bedone using soplit and apply rule
```

```python
import pandas as pd

data = {
    'Item': ['Cake', 'Cake', 'Bread', 'Pastry', 'Cake'],
    'Flavor': ['Chocolate', 'Vanilla', 'Whole Wheat', 'Strawberry', 'Chocolate'],
    'Price': [250, 220, 80, 120, 250]
}

df = pd.DataFrame(data)
print(df)
```

```python
# grouping data based on a single column
grouped=df.groupby("Item")
print(grouped)
# it doesn't show the result directly it just creates a grouped objects .
#To actually see the data we need to apply a method like sum(),mean()
```

```python
print(df.groupby("Item")["Price"].sum())
```

```python
# grouip by multiple columns 
print(df.groupby(["Item","Flavor"])["Price"].sum())
```

## different types of joins in pandas 

The pandas module contains various feature to perform various operations in dataframe like joins ,concatenate ,delete,add etc..five types of joins are :

Inner Join 

Left Outer Join


Right Outer Join 


Full outer Join



Index join

```python
# To understand different type of join we need to make two dataframes ,namely a and b
```

```python
# Data Frame A
import pandas as pd
a=pd.DataFrame()
d = {'id': [1, 2, 10, 12], 
     'val1': ['a', 'b', 'c', 'd']}
a=pd.DataFrame(d)
a
```

```python
# data Frame B 
import pandas as pd
d = {'id': [1, 2, 9, 8],
     'val1': ['p', 'q', 'r', 's']}
b=pd.DataFrame(d)
b
```

### pandas inner join

Inner join is the most common type of join .It retuirn the dataframe with common rows .This is similar to intersection of two sets 

```python
df=pd.merge(a,b,on="id",how="inner")
df
```

```python
# left outer join
df=pd.merge(a,b,on="id",how="left")
df
```

```python
# Right outer join 
df=pd.merge(a,b,on="id",how="right")
df
```

```python
# full outer join 
df=pd.merge(a,b,on="id",how="outer")
df
```

```python
# pandas index join
df=pd.merge(a,b,left_index=True,right_index=True)
df
```

```python
df=pd.read_csv("nba (1).csv")
aggregate_data=df.groupby(["Team","Position"]).agg(
    total_salary=("Salary","sum"),
    avg_salary=("Salary","mean"),
    player_count=("Name","count"))
display(aggregate_data)
```
