### Introduction to matplotlib


Matplotlib is a powerfull and versatile open-soyrse plotting library for python . It enables users to graphically represents data,facilitating easier analysis and understanding . If you want to convert your baring data into interactive plots and graphs ,matplotlib is the tool for you.




```python
import matplotlib.pyplot as plt
x=[0,1,2,3,4]
y=[0,1,4,8,1]

plt.plot(x,y)
plt.show()
```

### components or parts of matplotlib figure

Figure : The overarching container that holds all plots elements ,acting as the canvas for visualization

### MAtplotlib pyplot 


Pyplot is a module within matplotlib that provides a MATLAB like interface for making plots .

Steps to use pyplot


Import matplotlib : Start by importing matplotlib.pyplot as plt


Create the data : prepare your data in the form of lists and  arrays


plot data : use plt.plot() to create the plot 


customize plot : Add titles,labels and other elements using methods like plt.title(),plt,xlabel() and ylabel() 


Display plot : use plt.show() to display the plot

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots()
ax.plot(x,y,marker="o",label="Data Points")
ax.set_title("Basic componentys of matplotlib figure")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show()
```
