### Line chart in matplotlib in python


Matplotlib is a data visualization library in python > Line chart are used to represent the relationship between two X and Y on diffferent axis 

```python
# matplotlib simple line plot
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4])
y=x*2
plt.plot(x,y)
plt.show()
```

### fill the area between two lines 

Using pyplot.fill_between() function we can fill in the region between two line plots in the graph .This will help us in understanding the margin of data between two line plots 

```python
# alpha=0.5 is used to adjustbthe transparency 

import matplotlib.pyplot as plt 
import numpy as np
x=np.array([1,2,3,4])
y=x*2
plt.plot(x,y)
x1=[2,3,4,5]
y1=[1,2,3,4]
plt.plot(x1,y1)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("multuple plots ")
plt.fill_between(x,y,y1,color="green",alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x*2

plt.plot(x, y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

plt.plot(x, y1, '-.')
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')

plt.fill_between(x, y, y1, color='green', alpha=0.5)
plt.show()
```

### Bar plot in matplotlib 


A bar chart uses the rectangular bar to represents the data categories ,with bar length or height propotional to their values 

```python
import matplotlib.pyplot as plt
import numpy as np
fruit=["Apple","Banana","Cherry","Dates"]
sales=[400,350,200,100]

plt.bar(fruit,sales)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("fruit sales")
plt.show()
```

### The functions takes several parameters


x: The categories(e.g:fruit)
height : The corresponding values 
width : The width of the bar (default is 0.8)
bottom : The baseline for the bars (default 0)
align : how to align bars("center","edge")


```python
# syntax : plt.bar(x,height,width,bottom,align)
```

```python
import matplotlib.pyplot as plt
import numpy as np
fruit=["Apple","Banana","Cherry","Dates"]
sales=[400,350,200,100]

plt.bar(fruit,sales,color='voilet')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("fruit sales")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]

plt.bar(fruits, sales, color='violet')
plt.title('Fruit Sales')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.show()
```

```python
# creating horizantal bar chart 
import matplotlib.pyplot as plt
import numpy as np
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]
plt.barh(fruits,sales)
plt.show()
```
