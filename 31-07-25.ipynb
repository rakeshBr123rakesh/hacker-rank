### Types of data visualization 


Bar chart : rectangular bars and columns are used in bar charts to deposit data 


It is effective for comparing categories or discrete data points 



Follows the steps to use bar graph in you jupyter notebook: 

import the matplotlib module


take x-axis input as array 


take y-axis input in an array 


plot bar chart using bar() function .we can set other options like color,width,to customize the bar graph


set the title of your graph by using title() function 


show the graph using show() function 

```python
import matplotlib.pyplot as plt
x=[10,25,34,43,56]
y=[13,45,32,23,44]
plt.bar(x,y,color='pink',width=8)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,25,34,43,56]
y=[13,45,32,23,44]
plt.title('Bar Graph')
plt.bar(x, y, color='dodgerblue', width=5)
plt.show()
```

Pie chart in jupyter notebook


A pie chart displays data as circular graph divided into slices ,and each slice represents a proportion or percentage of the whole


follows the below steps to use pie chart in jupyter notebook


import the matplotlib module 


take the labels of your data in array



take the values in an array 


plot the pie chart in an array using pie() method 


set the title of your graph by using title() function 

show the graph by using title() function 


show graph using show() function

```python
import matplotlib.pyplot as plt 
x = [35, 20, 30, 40, 50, 30]
y = ['Apple', 'Bananna', 'Grapes', 'Orange', 'PineApple', 'Dragon Fruit']


plt.title("pie chart")
plt.pie(x,labels=y)
plt.show()
```

### Types of error bar 


vertical error bar : applied when the uncertyainity is along the Y-axis(deprndent variable) 


horizantal error bar : Used when uncertainity lies along the x-axis 


```python
# creating a simple graph
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[1,2,1,2,1,2]
plt.plot(x,y)
```

```python
# adding error to the y-values 

# this is for showing the uncertainity in the dependent variable 

import matplotlib.pyplot as plt
x =[1, 2, 3, 4, 5, 6, 7]
y =[1, 2, 1, 2, 1, 2, 1]

y_error=0.2

plt.plot(x,y)


plt.errorbar(x,y,yerr=y_error,fmt="o")
```

```python
# adding error to the x-axis 
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[1,2,1,2,1,2]

x_error=0.5
plt.plot(x,y)
plt.errorbar(x,y,xerr=x_error,fmt="o")
```

```python
# adding error bar to both x and y axis 

import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[1,2,1,2,1,2]
x_error=0.5
y_error=0.5
plt.plot(x,y)
plt.errorbar(x,y,xerr=x_error,yerr=y_error,fmt="o")
```

```python
# variable error in x and y 
# this demonstrate the how error bars can vary in length depending on ythe data ,reflecting different 
# level of uncertainity for each data points 

import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[1,2,1,2,1,2]

y_errormin=[0.1,0.2,0.9,0.8,0.2]
y_errormax=[0.2, 0.4, 0.6, 
             0.4, 0.2]


x_error=0.5
y_error=[y_errormin,y_errormax]


plt.errorbar(x,y,yerr=y_error,xerr=x_error,fmt="o")
```

```python
import matplotlib.pyplot as plt

x =[1, 2, 3, 4, 5]
y =[1, 2, 1, 2, 1]

# creating error
y_errormin =[0.1, 0.5, 0.9,
             0.1, 0.9]
y_errormax =[0.2, 0.4, 0.6, 
             0.4, 0.2]

x_error = 0.5
y_error =[y_errormin, y_errormax]

# plotting graph
# plt.plot(x, y)
plt.errorbar(x, y,
             yerr = y_error,
             xerr = x_error, 
             fmt ='o')
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(10)/10
y=(x+0.1)**2

y_err=np.linspace(0.05,0.2,10)
plt.plot(x,y)
plt.errorbar(x,y,yerr=y_err,fmt="o")
```
