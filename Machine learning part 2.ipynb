Predicting if a person would buy life insurence based on his age using logistic regression

Abov problem is binary logistic regression problem as there are only two possible outcomes (i.e. if person buys insurence or he/she doesn't)


```python
import pandas as pd
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
df=pd.read_csv("D:/career/sem 1/AI/AI_Data/insurance_data.csv")
df
```

```python
plt.scatter(df.age,df.bought_insurance,marker="+",color="red")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df[["age"]],df.bought_insurance,train_size=0.8)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
X_test
```

```python
y_predicted=model.predict(X_test)
```

```python
y_predicted
```

```python
model.predict_proba(X_test)
```

```python
model.score(X_train,y_train)
```

model.coef_ indicates the value of m in y=m*x+b equation
