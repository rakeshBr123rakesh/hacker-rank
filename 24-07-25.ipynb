```python
# pandas dtaframe.drop_duplicates()
#when we re working with data in pandas one common task is removing duplicates rows to ensure 
# clean and accurate datasets .the drop_duplicates() methos in pandas is designed to make this proces quick and easy
```

```python
import numpy as np
data = {
    "Name": ["Alice", "Bob", "Alice", "David"],
    "Age": [25, 25, 25, np.NaN],
    "City": ["NY", "LA", "NY", "Chicago"]}
df=pd.DataFrame(data)
print("origibal dataframe")
print(df)
```

```python
# modified duplictes
rm_null_df=df.drop_duplicates()
print(rm_null_df)
```

```python
# there is no duplicate in a dataframe
```

```python
# syntax for drop_duplicates()
# parameters :
# subset : Specifies the columns to check for duplicates .If not provided all columns are considered
```

```python
# 2. keep : finds which duplicates to keep :
 # 'first':keeps first occurences ,removes subsequent duplicates 
# 'last': keeps the first occurences ,removes subsequent duplicates 
# false : remove all occurences of duplicates 
```

```python
# inplace : If true it modifies the original DataFrame directly if false ,return new dataframe
```

```python
# dropping duplicates based on specific columns 
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'David'],
    'Age': [25, 30, 25, 40],
    'City': ['NY', 'LA', 'SF', 'Chicago']})
df_cleaned=df.drop_duplicates(subset=["Name"])
print(df_cleaned)
```

```python
# Dropthe last occurence of duplicates 
# By default drop_duplicates() retain the first occurance of duplicates .
#If we want the last occurance we can kkeep ="last"
```

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'David'],
    'Age': [25, 30, 25, 40],
    'City': ['NY', 'LA', 'NY', 'Chicago']
})

df_cleaned=df.drop_duplicates(keep="last")
print(df_cleaned)
```

```python
# If we want to remove the all rows that are duplicte and keep only unique
#rows then we can use keep=False
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'David'],
    'Age': [25, 30, 25, 40],
    'City': ['NY', 'LA', 'NY', 'Chicago']})
df_cleaned=df.drop_duplicates(keep=False)
print(df_cleaned)
```

```python
# Modifying the original dataframe directly
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'David'],
    'Age': [25, 30, 25, 40],
    'City': ['NY', 'LA', 'NY', 'Chicago']})
df.drop_duplicates(inplace=True)
print(df)
```

```python
# if we want to remove duplicates based on specified columns then 
df_cleaned=df.drop_duplicates(subset=["Name","City"])
print(df_cleaned)
```

```python
nan_value=float("NaN")
df.replace("",nan_value,inplace=True)
df.dropna(how="all",axis=1,inplace=True)
display(df)
```
