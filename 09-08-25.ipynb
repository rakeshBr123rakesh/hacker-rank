```python
# Matplotlib scatter 

# Scatter plots are one of the most fundamental and powerfull tools for visualizing relationship
# between two numerical variables
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([12, 45, 7, 32, 89, 54, 23, 67, 14, 91])
y = np.array([99, 31, 72, 56, 19, 88, 43, 61, 35, 77])

plt.scatter(x, y)
plt.title("Basic Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
# parameters in scatter plot matplotlib
```

x,y : sequence of data points


s : marker size


c : color of the marker


marker : shape of the marker


label : legend label for the dataset

```python
x1 = np.array([160, 165, 170, 175, 180, 185, 190, 195, 200, 205])
y1 = np.array([55, 58, 60, 62, 64, 66, 68, 70, 72, 74])

x2 = np.array([150, 155, 160, 165, 170, 175, 180, 195, 200, 205])
y2 = np.array([50, 52, 54, 56, 58, 64, 66, 68, 70, 72])

plt.scatter(x1,y1,color="blue",label="Group 1")
plt.scatter(x2,y2,color="pink",label="Group2")

plt.xlabel("X axis ")
plt.ylabel("Y axis ")
plt.title("comparision of width and height")
plt.legend()
plt.show()
```

```python
# thsi example is for demonstrate the how to customize  a scatter plot using different marker
# size and colors for each points transparency and edge color is also adjusted 

x = np.array([3, 12, 9, 20, 5, 18, 22, 11, 27, 16])
y = np.array([95, 55, 63, 77, 89, 50, 41, 70, 58, 83])

a = [20, 50, 100, 200, 500, 1000, 60, 90, 150, 300] # size
b = ['red', 'green', 'blue', 'purple', 'orange', 'black', 'pink', 'brown', 'yellow', 'cyan'] # color
plt.scatter(x,y,s=a,c=b,alpha=0.6,edgecolors="white",linewidth=1)
plt.title("scatter plot with varying color and size")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
sizes = [30, 80, 150, 200, 300]  # Bubble sizes
plt.scatter(x,y,s=sizes,alpha=0.6,edgecolors="pink",linewidth=1)
plt.title("Bubble plot example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# In this example we we map data values to clors using a colormap(cmap) and colorbar

x = np.random.randint(50, 150, 100)
y = np.random.randint(50, 150, 100)
colors = np.random.rand(100)  # Random float values for color mapping
sizes = 20 * np.random.randint(10, 100, 100)
plt.scatter(x,y,c=colors,s=sizes,cmap="viridis",alpha=0.7)
plt.colorbar(label="Color Scale")
plt.title("scatter plot with colormap and colorbar")
plt.show()
```

```python
# if we want to change the marker style then 
plt.scatter(x,y,s=100,color="pink",alpha=0.8,marker="^")
plt.show()
```

```python
# here X -axis represents the bin ranges and Y-axis represents the 
#information about frequency
```

```python
# creating the basic histogram in matplotlib 
import matplotlib.pyplot as plt
import numpy as np
data=np.random.randn(1000)
plt.hist(data,bins=30,color="skyblue",edgecolor="black")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("basic Histogram")
plt.show()
```

```python
# Creating the histogram with density plot
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
data=np.random.randn(1000)
sns.histplot(data,bins=30,kde=True,color="lightgreen",edgecolor="red")
plt.xlabel("Values")
plt.ylabel("Density")
plt.title("customized histogram with density plot")
plt.show()
```
