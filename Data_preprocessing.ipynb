operating on NULL values 
pandas treats None and NaN as esentialyy for indicating the missing and null values . To facilitate this convention,there are several usefull methods for detecting ,removingand replacing null values in pandas data structure

They are:

1.isnull():generate boolean mask indicating missing values
2.notnull(): opposite of isnull()
3.dropna(): return a filtered version of the data
4.fillna(): return a copy of the datawith missing values

Dropping the NULLvales:

1.dropna():(which removes NA values)
2.fillna():(which fills NA values)

we cannot drop single value from DataFrame we only drop full rows or columns.
By default dropna() will drop all rows in which any null values is present 

```python
df1=dataset.dropna()
df1
```

```python
df1.shape
```

Alternatively we can drop the NA values along with different axis .drop the all values containing null values

```python
df2=dataset.dropna(axis="columns")
df2
```

```python
#here axis=1 means columns
df3=dataset.dropna(axis=1)
df3
```

```python
dataset
```

but this drops some good data as well you might rether be interested in dropping rows or columns with all NA values,or majority of NA values
This can be specified through the how or thread parameters,which allow fine control of the nulls to allow through 

The default is how="any" such that any rows and columns containing a null values will be dropped .You can also specify how="all", which will drop rows and columns that are all null values 

```python
df4=dataset.dropna(axis="columns ",how="all")
df4
```

```python
print(dataset)
```

```python
df5=dataset.dropna(axis="rows",thresh=3)
df5
```

Filling null values

sometimes rather than dropping NA values rather replace them with a valid value. This might be a single number like zero,or it might be some sort of imputation or interpolation from the good values .  

```python
dataset
```

```python
df6=dataset.fillna(0)
df6
```

```python
df6=dataset.fillna(1)
```

```python
df6
```

we can specify a forward-fill to propagate the previous value forward 

```python
df8=dataset.fillna(method="ffill")
df8
```

```python
print(dataset)
```

```python
df9=dataset.fillna(method="bfill") 
df9
```

```python
print(dataset)
df10=dataset.fillna(method="ffill")
df10
```

if a previous value is not available during a forward fill or backward fill ,the NA value remains
